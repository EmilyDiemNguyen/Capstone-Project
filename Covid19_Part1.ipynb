## Data Cleaning

```python
data.info()
```

```python
# convert 'date' from object to datetime for better visualization
data['date'] = pd.to_datetime(data['date'])
```

```python
data.dtypes
```

```python
print('Continent:', data.continent.unique())
print('Number of continent:', data.continent.nunique())
```

```python
# look at 'location' attribute
print('Location:', data.location.unique())
print('Number of location:', data.location.nunique())
```

```python
# check 'location' attribute when 'continent' is null
print('Location when continent is null:', data[data.continent.isna()].location.unique())
```

When 'continent' is null, 'location' don't show countries but shows continents and some other groups. This can be considered as duplicate values. In this project, we will calculate some ratios on country base. Therefore, I decided to remove all records when 'continent' is null to keep 'location' only contains countries.

```python
data = data.dropna(subset=['continent'], axis = 0)
```

```python
# check 'location' attribute after removing records with 'continent' null values
print('Updated Location:', data.location.unique())
print('Number of updated location:', data.location.nunique())
```

```python
# Now 'location' only contains countries, so we will change attribute name from 'location' to 'country'
data.rename(columns = {'location':'country'}, inplace = True)
```

```python
data.country.value_counts()
```

```python
# check how many records left for further analysis
data.shape
```

**Find out what countries missing population data**

```python
# The data of 'population' has to be available for further analysis, because we will calculate ratios based on population of each country. First find out what countries missing 'population' value
print('Country with null population:', data[data.population.isna()].country.unique())
```

```python
# let check population input for all records belongs to 'Northern Cyprus'
data.groupby('country')['population'].get_group('Northern Cyprus').sum()
```

```python
#Check other attributes when 'country' is 'Northern Cyprus'
data[data.country.str.contains('Northern Cyprus')].isna().sum()
```

```python
# Most of records belonging to 'Northern Cyprus' is null, we will remove all records of 'Northern Cyprus'
data = data[data.country != 'Northern Cyprus']
data.shape
```

```python
# Check how many countries left for further analysis
print('Number of countries:', data.country.nunique())
```

**Deal with null values (NA)**

```python
#Check missing values
data.isna().sum().sort_values(ascending=False)
```

For cumulative data: 'total_cases', 'total_deaths', 'people_fully_vaccinated', 'icu_patients'

```python
# Fill NA by ffill() function ('forward fill'), then we will fill the rest by 0
# Null values will be replaced by it’s previous value in the column, if there is no previous value, NA will be replaced by 0
data[['total_cases', 'total_deaths','people_fully_vaccinated','icu_patients']] = data.groupby('country')[['total_cases', 'total_deaths','people_fully_vaccinated','icu_patients']].fillna(method='ffill').fillna(0)
```

For non-cumulative data: 'new_cases', 'new_deaths'

```python
# Fill NA by the difference between 2 records then fill the rest by 0
data.new_cases = data.groupby('country').new_cases.fillna(data.groupby('country').total_cases.diff(periods=1)).fillna(0)
data.new_deaths = data.groupby('country').new_deaths.fillna(data.groupby('country').total_deaths.diff(periods=1)).fillna(0)
```

```python
# check data set after dealing with NA
data.isna().sum().sum()
```
